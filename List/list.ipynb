```python
#filtering
n = [10, 12,2.3,69.3,"jahid", 2+3j, False, "Rayhan" ,45, True]
int_l = []
f_l = []
b_l = []
s_l = []
c_l = []

for i in n :
  if type(i) == int:
    int_l.append(i)
  elif type(i) == float:
    f_l.append(i)
  elif type(i) == bool:
    b_l.append(i)
  elif type(i) == str:
    s_l.append(i)
  else:
    c_l.append(i)

print(int_l)
print(f_l)
print(b_l)
print(s_l)
print(c_l)
```
