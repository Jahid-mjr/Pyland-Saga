```python
def av(*num):
  a = 0
  b = 0
  for i in num:
    b = b + i
    a = a + 1

  return b/a

av(12,12,12,12,12)

```

```python
def dic(**a):
  print(a)

dic( a = 45, b = 45)
dic( a = "jahid", b = "haj")
```

```python
def dic(**a):
  print(a.keys())

dic( a = 45, b = 45)
dic( a = "jahid", b = "haj")
```

```python
def dic(**a):
  print(a.values())

dic( a = 45, b = 45)
dic( a = "jahid", b = "haj")
```

```python
def dic(**a):
  print(a.items())

dic( a = 45, b = 45)
dic( a = "jahid", b = "haj")
```

```python
m = [50,60,70,80,90,100]

new_m = []

for i in m:
  if i == 100:
    new_m.append(i)
  else :
    new_m.append(i+3)

print(new_m)
```

```python
#new_mm = [i+3 for i in m if i != 100]

new_mm = [i+3 if i != 100 else i for i in m]

new_mm
```
